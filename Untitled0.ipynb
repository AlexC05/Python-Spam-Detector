```python
from unsloth import FastLanguageModel
from datasets import load_dataset
from trl import SFTTrainer
from transformers import TrainingArguments

import os
os.environ["WANDB_DISABLED"] = "true"

# ============================
# 1. Load dataset
# ============================
dataset = load_dataset("json", data_files={
    "train": "train.jsonl",
    "validation": "valid.jsonl",
})

# ============================
# 2. Clean dataset
# ============================
def is_valid(example):
    return ("prompt" in example and
            "response" in example and
            isinstance(example["prompt"], str) and
            isinstance(example["response"], str) and
            example["prompt"].strip() != "" and
            example["response"].strip() != "")

dataset["train"] = dataset["train"].filter(is_valid)
dataset["validation"] = dataset["validation"].filter(is_valid)

# ============================
# 3. Load model Gemma 1B
# ============================
model, tokenizer = FastLanguageModel.from_pretrained(
    "google/gemma-3-1b-it",
    load_in_4bit=True,
)

model = FastLanguageModel.get_peft_model(
    model,
    r=8,
    lora_alpha=32,
    lora_dropout=0,
    target_modules=["q_proj", "v_proj"],
)

# ============================
# 4. Format function
# ===========================

def formatting_func(example):
    prompts = example["prompt"]
    responses = example["response"]
    formatted_list = []

    for p, r in zip(prompts, responses):
        text = f"<prompt>\n{p}\n<response>\n{r}"
        formatted_list.append(text)

    return formatted_list


# ============================
# 5. Training arguments
# ============================
training_args = TrainingArguments(
    output_dir="gemma1b_spam_lora",
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=16,
    logging_steps=10,
    learning_rate=2e-4,
    num_train_epochs=3,
    fp16=True,
)

# ============================
# 6. Trainer
# ============================
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset["train"],
    eval_dataset=dataset["validation"],
    formatting_func=formatting_func,
    args=training_args,
    max_seq_length=1024,
)

trainer.train()

trainer.save_model("gemma1b_spam_lora")
tokenizer.save_pretrained("gemma1b_spam_lora")

print("FINETUNING COMPLET!")
```
